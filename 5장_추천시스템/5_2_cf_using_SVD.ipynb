```python
movie_info_df = pd.read_csv('/content/drive/MyDrive/수업자료/00_dataset/movielens/ml-100k/u.item', sep='|', encoding='latin')
movie_info_df.columns = ['movieID' , 'movie_title' , 'release_date' , 'video_release_date' ,
              'IMDb_URL' , 'unknown' , 'Action' , 'Adventure' , 'Animation' ,
              'Children' , 'Comedy' , 'Crime' , 'Documentary' , 'Drama' , 'Fantasy' ,
              'Film-Noir' , 'Horror' , 'Musical' , 'Mystery' , 'Romance' , 'Sci-Fi' ,
              'Thriller' , 'War' , 'Western']
movie_info_df
```

```python
movie_rating_df = pd.merge(rating_df, movie_info_df[['movieID', 'movie_title']], on = 'movieID')
movie_rating_df['rating']= movie_rating_df['rating'].astype(float)
movie_rating_df
```

```python
user_ids = movie_rating_df["userID"].unique().tolist()

user2idx_dict = {x: i for i, x in enumerate(user_ids)}
idx2user_dict = {i: x for i, x in enumerate(user_ids)}

movie_ids = movie_rating_df["movieID"].unique().tolist()

movie2idx_dict = {x: i for i, x in enumerate(movie_ids)}
idx2movie_dict = {i: x for i, x in enumerate(movie_ids)}

movie_rating_df["user_idx"] = movie_rating_df["userID"].map(user2idx_dict)
movie_rating_df["movie_idx"] = movie_rating_df["movieID"].map(movie2idx_dict)

num_users = movie_rating_df["userID"].nunique()
num_movies = movie_rating_df["movieID"].nunique()
print(num_users, num_movies)

movie_rating_df = movie_rating_df.sort_values(['user_idx', 'movie_idx'])
movie_rating_df = movie_rating_df.reset_index(drop=True)
movie_rating_df
```

```python
from scipy.sparse import coo_matrix, linalg

rating_sparse_arr = coo_matrix((movie_rating_df['rating'], (movie_rating_df['user_idx'], movie_rating_df['movie_idx'])), shape = (num_users, num_movies))
rating_sparse_arr
```

### SVD를 사용해 평점 예측
- SVD는 linalg
- SVDs는 SVD의 sparse matrix version

```python
# SVD는 분해된 3 행렬이 return 됩니다.
U, S, V = linalg.svds(rating_sparse_arr, k=250)
```

```python
print(U.shape, S.shape, V.shape)
```

```python
test_user_id = 30
test_user_idx = movie_rating_df[movie_rating_df['userID'] == test_user_id].user_idx.iloc[0] 
test_user_idx
```

```python
# 세 행렬을 다시 곱하면 원래의 행렬로 복원됩니다. --> 이 과정에서 숨겨진 평점이 예측됨
# 모든 행렬을 복원할 필요 없기 때문에 target user data만 입력
# S는 대각성분만 있기 때문에 대각행렬로 돌려줍니다.

pred = U[test_user_idx] @ np.diag(S) @ V
pred.shape
```
