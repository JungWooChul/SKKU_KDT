```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%ED%94%84%EB%A1%9C%EC%95%BC%EA%B5%AC'
resp = requests.get(url)
```

```python
# 200 : 성공
# 404 : 실패
resp
```

```python
# bs 형태로 변환
resp.content
```

```python
soup = BeautifulSoup(resp.content)
a_tags = soup.select('a.news_tit')
len(a_tags)
```

```python
a_tags[0]
```

- tag의 속성 파싱하기 : 속성은 딕셔너리의 key값 처럼 지정해주면 된다.

```python
for i in a_tags:
    print(i['title'])
```

```python
dic_news = {'제목':[], '링크':[]}
for i in a_tags:
    dic_news['제목'].append(i['title'])
    dic_news['링크'].append(i['href'])
```

```python
import pandas as pd
df_news = pd.DataFrame(dic_news)
df_news.to_csv('naver_news_article.csv', encoding='utf-8-sig', index=False)
```

- 연관 검색어 키워드 가져오기

```python
# class가 'tit'인 'div' 태그를 가진 다른 것들이 있을거같다는 불안감
soup = BeautifulSoup(resp.content)
div_tags = soup.select('div.tit')
div_tags = soup.select('a.keyword div.tit') # 공백은 상위 태그와 하위태그를 구분해주는 구분자
len(div_tags) # 길이를 확인
```
