### 문제 11
다음과 같이 4개의 성적이 담긴 리스트 score가 있다.

4개 성적의 평균이 60점 이상이고, 각 점수에 40점 미만의 점수가 없어야 '합격'일 때,

합격인지 불합격인지 출력하는 코드를 작성하세요.

```python
score = [55, 90, 80, 35]
mean = sum(score) / len(score)

# 더 효율적인 방법은 없을지 고민
check = 0
for i in score:
    if i <=40:
        check = 1

if mean>=0 and check !=1:
    print('합격')
else:
    print('불합격')
```

### 문제12
다트를 던져서 50점과 가까울수록 높은 점수를 매긴다 할 때,

45점 이상 55점 미만일 때 '+10점',

40점 이상 60 미만일 때 '+7점',

30점 이상 70점 미만일 때 '+5점'을 출력하는 코드를 작성하세요.

```python
score = 43
if 45<=score<=55:
    print('+10점')
elif 40<=score<=60:
    print('+7점')
elif 30<=score<=70:
    print('+5점')
```

### 문제2
사과 2개, 망고 3개, 수박 1개를 사려고 한다.

살 수 없는 과일을 출력하고, 살 수 있는 과일은 구입할 갯수와 가격을 계산해 총 가격을 출력해보세요.(코드 형식 무관)

```python
fruit = {'사과' : '5000', '바나나' : '7000', '망고' : '8000'}
price = 0

wanted = {'사과' : 2, '망고' : 3, '수박' : 1}
print('살 수 없는 과일 : ',end="")
for i in list(wanted.keys()):
    if i in fruit:
        price += wanted[i] * int(fruit[i])
    else:
        print(i)

print('총 가격 : ', price)
        
```
