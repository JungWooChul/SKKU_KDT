# 다음 레이어를 활용해 아래 문제들을 풀어보세요

아래 문제들은 실용성은 없고 빌드 연습을 위해 임의로 꼬아놓은 모델입니다.

## 사용 레이어 및 활용법
(실수로 빠진게 있을 수 있습니다.;;)
- Dense()(x)
- Concatenate()([x1, x2])
- Multiply()([x1, x2])
- Add()([x1, x2])
- Subtract()([x1, x2])

## 문제 1

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Add, Multiply
from tensorflow.keras.models import Model
```

```python
## 모델을 구축해보세요
input_1eft = tf.keras.Input(shape = (128,))
input_right = tf.keras.Input(shape = (128,))

dense_left_output = Dense(64, activation = 'relu')(input_1eft)
dense_right_output = Dense(64, activation = 'relu')(input_right)

x_left = Add()([dense_left_output, dense_right_output])
x_right = Multiply()([dense_left_output, dense_right_output])

x = tf.keras.layers.concatenate([x_left, x_right], axis=1)
output_layer = Dense(1, activation = 'sigmoid')(x)
```

```python
model_simple = Model(inputs=[input_1eft, input_right], outputs=output_layer)
tf.keras.utils.plot_model(model_simple, show_shapes=True)
```

## 문제 2

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers import Input, Dense, concatenate
from tensorflow.keras.models import Model
```

```python
## 모델을 구축해보세요
input_layer = tf.keras.Input(shape = (128,))

x_dense1 = Dense(64)(input_layer)

x_dense2 = Dense(64)(x_dense1)

x_concat1 = concatenate([x_dense1, x_dense2], axis = 1)
x_mul = Multiply()([x_dense1, x_dense2])

x_concat2 = concatenate([x_mul, x_dense1], axis = 1)

x_add = Add()([x_concat1, x_concat2])

output_layer = Dense(1)(x_add)
```

```python
model_2 = Model(inputs=input_layer, outputs=output_layer)
tf.keras.utils.plot_model(model_2, show_shapes=True)
```

## 문제 3

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, concatenate, Subtract, Add
from tensorflow.keras.models import Model
```

```python
## 모델을 구축해보세요
input_a = tf.keras.Input(shape = (128,))
input_b = tf.keras.Input(shape = (128,))

a = Dense(64)(input_a)
b = Dense(64)(input_b)

a_dense = Dense(64)(a)
sub1 = Subtract()([a,b])
b_dense = Dense(64)(b)

add1 = Add()([a_dense, b_dense])
concat = concatenate([add1, sub1], axis = 1)

output_layer_1 = Dense(1)(concat)
output_layer_2 = Dense(10)(concat)
```

```python
model_3 = Model(inputs=[input_a, input_b], outputs=[output_layer_1, output_layer_2])
tf.keras.utils.plot_model(model_3, show_shapes=True)
```
